```python
from bgn import BigramNetwork
```

```python
network = BigramNetwork("data/cities_normalized_hungary.txt", 100)
```

```python
k = 120
for i in range(k):
    l = network.train()
    if (i<5 or k-5 <= i):
        print(f'{i:3}', l)
```

```python
# Predict a city name
for i in range(20):
    letters:list[str] = ['<b>']
    while True:
        sampled = network.sample(network.stoi[letters[-1]])
        letters.append(network.itos[int(sampled)])
        if (letters[-1] == '<b>'):
            break

    print(''.join(letters))
```
